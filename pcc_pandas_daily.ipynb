# Reading RADOLAN RX and RY filterd products
-----------------------------------------

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import wradlib as wrl
% matplotlib inline
```

# Read CSV with pandas
------------------

```python
# read CSV and create a data frame with time index from Unnamed: 0
dfr = pd.read_csv('/automount/ags/velibor/data/radolan/rxy_20090101000000-20091231235000.csv')
dfrr = dfr.set_index(pd.DatetimeIndex(dfr[u'Unnamed: 0']))

# Drop Unnamed: 0 Column
dfrr = dfrr.drop([u'Unnamed: 0'],axis=1)


```

# View data set in taublar and as statistics
------------------------------------------


```python
ma = dfrr[u'rsum'].groupby(dfrr[u'rsum'].index.hour).mean()
mad = dfrr[u'rsum'].groupby(dfrr[u'rsum'].index.hour).median()

mstd = dfrr[u'rsum'].groupby(dfrr[u'rsum'].index.hour).std()
mmin = dfrr[u'rsum'].groupby(dfrr[u'rsum'].index.hour).min()
mmax = dfrr[u'rsum'].groupby(dfrr[u'rsum'].index.hour).max()


plt.plot(ma.index, ma, 'k', label='mean')
plt.plot(ma.index, mad, 'b', label='median')
plt.legend()
#plt.fill_between(mstd.index, ma-2*mstd, ma+2*mstd, color='b', alpha=0.2)
plt.grid()
plt.xlabel('hour')
plt.ylabel('rainsum')
plt.xlim(0,23)
```

```python
ma = dfrr[u'rsum'].groupby(dfrr[u'rsum'].index.month).mean()



plt.plot(ma.index, ma, 'k')
#plt.fill_between(mstd.index, ma-2*mstd, ma+2*mstd, color='b', alpha=0.2)
plt.grid()
plt.xlabel('time')
plt.ylabel('rainsum')
```
